### 3.1. Player Individual Stats

#### 3.1.1. Throwing Accuracy per thrower

```python
df_throwing = df_passes.groupby(['throwerUUID']).aggregate({'isThrowerError': ['count', 'sum']}).reset_index()
df_throwing['throwing_perc'] = 1 - df_throwing[('isThrowerError', 'sum')] / df_throwing[('isThrowerError', 'count')]
df_throwing.columns = ['_'.join(col).strip() if col[1] else col[0] for col in df_throwing.columns.values]
# df_throwing = pd.merge(df_throwing, df_players[['uuid', 'name']], left_on=['throwerUUID'], right_on=['uuid'])

# renaming cols
df_throwing = df_throwing.rename(columns={
    'throwerUUID': 'uuid',
    'isThrowerError_count': 'throws', 
    'isThrowerError_sum': 'throw_failed',
    'throwing_perc': 'throwing_perc',
})
```

```python
df_throwing.head()
```

#### 3.1.2 Receiver Catching Percentage

```python
df_catching = df_passes.groupby(['receiverUUID']).aggregate({'isReceiverError': ['count', 'sum']}).reset_index()
df_catching['catching_perc'] = 1 - df_catching[('isReceiverError', 'sum')] / df_catching[('isReceiverError', 'count')]
df_catching.columns = ['_'.join(col).strip() if col[1] else col[0] for col in df_catching.columns.values]
# df_catching = pd.merge(df_catching, df_players[['uuid', 'name']], left_on=['receiverUUID'], right_on=['uuid'])

# renaming cols
df_catching = df_catching.rename(columns={
    'receiverUUID': 'uuid', 
    'isReceiverError_count': 'catches',
    'isReceiverError_sum': 'drops',
    'catching_perc': 'catching_perc'
})
```

```python
df_catching.head()
```

#### 3.1.3. Goals, Assists, Hockey Assist

```python
# isAssist
is_scoring_pass_mask = df_passes['isAssist'] == True
df_scoring_points = df_passes[is_scoring_pass_mask] 
df_assists = df_scoring_points.groupby(['throwerUUID'])['isAssist'].sum().reset_index()
df_goals = df_scoring_points.groupby(['receiverUUID'])['isAssist'].sum().reset_index()

# isSecondaryAssist
df_hockey_assists = df_passes.groupby(['throwerUUID'])['isSecondaryAssist'].sum().reset_index()

# rename columns
df_assists.columns = ['uuid', 'assists']
df_goals.columns = ['uuid', 'goals']
df_hockey_assists.columns = ['uuid', 'hockey_assists']
```

```python
df_players_stats = df_players[['uuid', 'name']]
for df in [df_assists, df_goals, df_hockey_assists]:
    df_players_stats = pd.merge(df_players_stats, df, on='uuid', how='outer')
df_players_stats = df_players_stats.fillna(0)
```

#### 3.1.4. Combining dataframes

```python
df_players_stats = df_players[['uuid', 'name']]
for df in [df_assists, df_goals, df_hockey_assists, df_throwing, df_catching]:
    df_players_stats = pd.merge(df_players_stats, df, on='uuid', how='outer')
df_players_stats = df_players_stats.fillna(0)
```

```python
idx_to_player_id_dct = {i: player_id for i, player_id in enumerate(df_players['uuid'])}
player_id_to_idx_dct = {player_id: i for i, player_id in enumerate(df_players['uuid'])}
```

```python
# create adjancy matrix for (1) successful_throw to (2) failed # 
num_players = len(df_players)
matrix_success = np.zeros((num_players, num_players))
matrix_failures = np.zeros((num_players, num_players))
matrix_proba = np.zeros((num_players, num_players))

for player_idx, player_uuid in enumerate(df_players['uuid']):
    # 1. compute receiver success proba
    t = df_passes[df_passes['throwerUUID'] == player_uuid]
    t['isError'] = t['isReceiverError'] | t['isThrowerError']
    t['isSuccess'] = ~t['isError']
    t1 = t.groupby('receiverUUID').aggregate({ 'isSuccess': ['sum'], 'isError': ['sum', 'count'],}).reset_index()
    t1.columns = ['receiverUUID', 'success', 'error', 'total', ]
    t1['perc'] = t1['success'] / t1['total']

    # 2. fill in matrices
    for _, row in t1.iterrows():
        receiver_idx = player_id_to_idx_dct[row['receiverUUID']]
        matrix_success[player_idx, receiver_idx] = row['success']
        matrix_failures[player_idx, receiver_idx] = row['error']
        matrix_proba[player_idx, receiver_idx] = row['success'] / row['total']
```
