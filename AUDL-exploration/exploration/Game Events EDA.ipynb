### What throw do players prefer?

```python
# get all events of type t:20
events_t20 = []
for row in events['homeEvents']:
    for event in row['events']:
        t = event['t']
        if t == 20:
            events_t20.append(event)
events_t20 = pd.json_normalize(events_t20)

# calculate average throw (x,y) coordinates and count num throws
aggregate_dict = {
    'x': 'mean',
    'y': 'mean',
    't': 'count'
}
average_player_throws = events_t20.groupby(['r']).aggregate(aggregate_dict)

# map receiver id to ext_player_id
average_player_throws = average_player_throws.reset_index()

average_player_throws['ext_player_id'] =  average_player_throws['r'].apply(
    lambda x: players[players['id'] == int(x)]['player.ext_player_id'].tolist()[0])

```
