```python
import pyarrow
import duckdb
from google.cloud import bigquery

client = bigquery.Client()
query = "SELECT * from `relax-melodies-android.ua_dashboard_prod.final_table` WHERE date >= '2025-01-01'"

# raw_table = client.query(query=query).to_arrow()
# duckdb.sql("CREATE TABLE final_table as SELECT * FROM raw_table")

df = client.query(query=query).to_dataframe()
```

```python
print(f"min date: {df['date'].min()}")
print(f"max date: {df['date'].max()}")
print(f"Unique network: {df['network'].unique()}")

MAIN_NETWORKS = ['Apple Search Ads', 'Facebook Ads', 'googleadwords_int', 'tiktokglobal_int', 
                 'tatari_linear', 'tatari_streaming']
                
# past covariates: measurements
TARGET = 'paid'
METRICS = ['cost_usd', 'impressions', 'clicks', 'installs', 'trials', 'paid', 'revenue', 'refund']
PAST_COVARIATES = []

# future covariates: calendar, weather forecasts, spend
FUTURE_COVARIATES = ['cost_usd', 'impressions', 'clicks', 'installs', 'trials']


```

```python
mask_android = df['platform'] == 'android'
mask_us = df['country'] == 'US'
mask_main_networks = df['network'].isin(MAIN_NETWORKS)
df_android = df[mask_android & mask_main_networks & mask_us].groupby(['date', 'network', 'platform', 'country']).sum()[METRICS].reset_index()
```
