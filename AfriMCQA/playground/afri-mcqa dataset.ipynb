## make prompt

```python
def get_user_prompt(sample, prompt_lang: str, qa_lang: str, answer_lang: str):
    """
    LANG = ["English", ...]
    """
    user_prompt = f""" 
    LANGUAGE: Respond in {answer_lang} only
    Question: {sample['eng_question'] if prompt_lang == 'English' else sample['native_question']}
    Choices:
    A. {sample['correct_en'] if qa_lang == 'English' else sample['correct_native']}
    B. {sample['wrong_en_o1'] if qa_lang == 'English' else sample['wrong_native_o1']}
    C. {sample['wrong_en_o2'] if qa_lang == 'English' else sample['wrong_native_o2']}
    D. {sample['wrong_en_o3'] if qa_lang == 'English' else sample['wrong_native_o3']}
    Answer:
    Short Explanation (30 words max):
    """
    return user_prompt
    
```

```python
lang = 'Akan_Twi'
sample = datasets[lang][0]
get_user_prompt(sample=sample, prompt_lang='English', qa_lang='English', answer_lang='English')
get_user_prompt(sample=sample, prompt_lang='English', qa_lang=lang, answer_lang='English')
get_user_prompt(sample=sample, prompt_lang=lang, qa_lang=lang, answer_lang='English')

    
```
