```python
import rustac
import pystac
from odc.stac import load

url = "https://data.source.coop/ausantarctic/ghrsst-mur-v2/ghrsst-mur-v2.parquet"

year = 2025

dicts = await rustac.read(url)
items = [pystac.Item.from_dict(d) for d in dicts["features"]]
```

```python
less_items = [i for i in items if i.datetime.year == year]

print(f"Found {len(items)} items, filtered to {len(less_items)} items for {year}")

data1 = load(
    less_items,
    chunks={"longitude": 4000, "latitude": 4000, "time": 1},
    measurements=["analysed_sst"],
    fail_on_error=False
)

data2 = load(
    items,
    chunks={"longitude": 4000, "latitude": 4000, "time": 1, "year": 2025}, # doesn't filter dates
    measurements=["analysed_sst"],
    fail_on_error=False
)
```

### Filtering using dates

```python
from datetime import datetime, timezone

start_date, end_date = "2025-01-01", "2026-01-01"
year, month, day, hour, second = 2025, 6, 23, 5, 30

# less_items = [i for i in items if i.datetime.year == year] # works!
# less_items = [i for i in items if i.datetime.month == month] # works!
# less_items = [i for i in items if i.datetime.day == day] # works!

format_str = "%Y-%m-%d"

# note: pystac use utc time
start_dt = datetime.strptime(start_date, format_str).replace(tzinfo=timezone.utc)
end_dt = datetime.strptime(end_date, format_str).replace(tzinfo=timezone.utc)

less_items = [i for i in items if start_dt <= i.datetime <= end_dt]
```
