```python
import os
import pandas as pd

from pyairtable import Api
from dotenv import load_dotenv
```

```python

load_dotenv()
AIRTABLE_TOKEN = os.getenv("AIRTABLE_TOKEN")
AIRTABLE_BASE_ID = os.getenv("AIRTABLE_BASE_ID")
AIRTABLE_TABLE_ID = os.getenv("AIRTABLE_TABLE_ID")
```

```python
api = Api(AIRTABLE_TOKEN)
```

```python
table = api.table(AIRTABLE_BASE_ID, AIRTABLE_TABLE_ID)
```

```python
t = table.all()
```

```python
contents = [c["fields"] for c in table.all()]
```

```python
df = pd.DataFrame.from_records(contents)
```

### Download Images

```python
cdn_images = [path for path in df['Image Download'].unique() if 'cdn' in str(path)]
```

```python
len(cdn_images)
```

```python
import requests
import os

def download_image_from_link(url: str, download_path: str):
    print(f"Downloading from url {url} ...}")
    image_path = url.split('/')[-1]
    full_path = os.path.join(download_path, image_path)
    img_data = requests.get(url).content
    with open(full_path, 'wb') as handler:
        handler.write(img_data)
```

```python
download_path = "/Users/emulie/Data/GuidedContentImages/"
```

```python
for url in cdn_images:
    download_image_from_link(url, download_path)
```
